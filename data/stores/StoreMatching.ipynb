```python
import pandas as pd
```

```python
data_main = pd.read_excel("aggregated.xlsx", "Main")
data_coop = pd.read_excel("aggregated.xlsx", "Coop")
data_migros = pd.read_excel("aggregated.xlsx", "Migros")
data_denner = pd.read_excel("aggregated.xlsx", "Denner")
data_spar = pd.read_excel("aggregated.xlsx", "Spar")
```

```python
data_main
```

```python
data_coop
```

```python
from scipy.spatial.distance import squareform, cdist
```

## Coop

```python
coop = pd.DataFrame(cdist(list(zip(data_coop.place_latitude, data_coop.place_longitude)),list(zip(data_main.LAT_CNTR, data_main.LONG_CNTR))),index=data_coop.index,columns=data_main.index)
coop['lowest_col'] = coop.idxmin(axis=1)
coop = coop.merge(data_main.GMDNAME, how="left", left_on="lowest_col", right_on=data_main.index)
data_coop = data_coop.merge(coop.GMDNAME, how="left", left_index=True, right_index=True)
```

```python
data_coop
```

```python
data_coop[data_coop.Gemeindename == data_coop.GMDNAME].count()
```

```python
data_coop.to_csv("coop_stores_aggregated.csv", index=False)
```
