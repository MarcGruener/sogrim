```python
import pandas as pd

agg = pd.read_excel("aggregated.xlsx", "Main")
predictions = pd.read_csv("predictions.csv")
data = pd.read_csv("data.csv")
```

```python
predictions = predictions.merge(agg[["GMDNAME", "LAT_CNTR", "LONG_CNTR"]], how='left', left_index=True, right_index=True)
data = data.merge(agg.GMDNAME, how='left', left_index=True, right_index=True)
```

```python
agg.to_csv('aggregated.csv', index=False)
predictions.to_csv('predictions.csv', index=False)
data.to_csv('data.csv', index=False)
```

```python
choice_model = "knnModel"
choice_option = "Consolidation"

if choice_option == "Consolidation":
  location_data = predictions[predictions[choice_model] < predictions.ANZAHL_FILIALEN_MIGROS]
elif choice_option == "Perfect":
  location_data = predictions[predictions[choice_model] == predictions.ANZAHL_FILIALEN_MIGROS]
elif choice_option == "Opportunities":
  location_data = predictions[predictions[choice_model] > predictions.ANZAHL_FILIALEN_MIGROS]
```
